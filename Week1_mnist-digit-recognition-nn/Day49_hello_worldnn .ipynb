# Day 49: Hello World Neural Network
**Goal:** Predict a simple numeric pattern using a neural network (y = 2*x)
**Platform:** Google Colab
**Libraries:** TensorFlow, Keras, NumPy


## Step 1: Import Libraries
We import the libraries needed for this project:
- `tensorflow` for deep learning
- `Sequential` and `Dense` for building the model
- `numpy` for numerical operations


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
```

## Step 2: Prepare Data
We create simple input (X) and output (y) arrays.
- X = input numbers
- y = target numbers following the pattern y = 2*x


```python
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)
```

## Step 3: Build the Model
- Use a Sequential model (stacked layers)
- Dense layer with 1 neuron (simple output)
- Input shape = 1 (one input feature)


```python
model = Sequential([
    Dense(1, input_shape=[1])
])
```

## Step 4: Compile the Model
- Optimizer: Stochastic Gradient Descent (sgd)
- Loss: Mean Squared Error (MSE) to measure error


```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Step 5: Train the Model
- Train for 200 epochs (repeated practice)
- The model adjusts weights to minimize loss


```python
history = model.fit(X, y, epochs=200, verbose=1)
```

## Step 6: Visualize Loss During Training
We plot the loss to see how the model improves over epochs.


```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss During Training')
plt.show()
```

## Step 7: Make Predictions
Test the model with new inputs to see if it learned the pattern y = 2*x


```python
print("Prediction for input 6:", model.predict(np.array([6], dtype=float)))
print("Prediction for input 10:", model.predict(np.array([10], dtype=float)))
```

## Step 8: Inspect Weights and Bias
Check the learned weight and bias values.
- Weight should be ~2
- Bias should be ~0


```python
weights, biases = model.layers[0].get_weights()
print("Weights:", weights)
print("Biases:", biases)
```

## Step 9: Key Learnings
- Neural Network can learn numeric patterns automatically
- Training adjusts weights and bias to reduce error
- Loss curve shows learning progress
- Predictions use learned weights + bias
- Even a tiny network can solve simple problems
